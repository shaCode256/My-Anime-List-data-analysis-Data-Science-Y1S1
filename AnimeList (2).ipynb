```python
#cleaning the data, let's drop columns which we think do not contribute our mission

a=animeFiltered.drop(['producer'], axis=1)
b=a.drop(['aired'], axis=1)
#animeFiltered.drop(['scored by'], axis=1)
c=b.drop(['related'], axis=1)
d=c.drop(['licensor'], axis=1)
e=d.drop(['ending_theme'], axis=1)
f=e.drop(['opening_theme'], axis=1)
g=f.drop(['aired_string'], axis=1)
h=g.drop(['title_english'], axis=1)
i=h.drop(['title_japanese'], axis=1)
j=i.drop(['anime_id'], axis=1)
k=j.drop(['title'], axis=1)
m=k.drop(['image_url'], axis=1)
n=m.drop(['background'], axis=1)
l=n.drop(['premiered'], axis=1)
o=l.drop(['title_synonyms'], axis=1)
p=o.drop(['rank'], axis=1)
q=p.drop(['airing'], axis=1)
r=q.drop(['scored_by'], axis=1)

clean=r
clean
```

```python
#Define our mission!
#the aim of this project is to forecast anime's success by it's score.

clean.loc[clean['score'] <7, 'score'] = 0
clean.loc[clean['score'] >= 7, 'score'] = 1

clean
```

```python
#duriation processing-

clean['duration'] = animeFiltered['duration'].str.replace(r'\D', '')
clean['duration'] = animeFiltered['duration'].str.replace(r'. per ep.', '')
clean['duration'] = clean['duration'].str.replace(r'.', '')

#deal with Unknowns
#get the most common one:
clean.duration #or add

clean.loc[clean['duration'] == 'Unknown', ['duration']] = '24 min'

#covert to timedelta-
clean['duration'] = pd.to_timedelta(clean['duration'])

#timedelta to minutes
clean['duration'] = clean['duration'].dt.total_seconds() / (60)


clean
```

```python
#duration processing 2-
#because 32/31 mins are the same in our matter, we'll divide the mins for categories:
#0: less than 23-short
#1: 23-45 medium
#2: more than 45- long

clean.loc[clean['duration'] < 23, 'duration'] = 0
clean.loc[clean['duration'] >= 23, 'duration'] = 1
clean.loc[clean['duration'] > 45, 'duration'] = 2

clean
```

```python
#cleaning null to spot and fill nulls
#dummies_replacement

obj_df = clean.select_dtypes(include=['object'])
obj_df.head()

#clean
obj_df[obj_df.isnull().any(axis=1)]

obj_df["broadcast"].value_counts #to see which is the most common one, so we'll fill the null ones with it
obj_df = obj_df.fillna({"broadcast": "Unknown"})

obj_df["studio"].value_counts #to see which is the most common one, so we'll fill the null ones with it
obj_df = obj_df.fillna({"studio": "Unknown"})

#get which duration is the most common
clean.duration.mode()
clean['duration']=clean['duration'].replace("Unknown", "24 min")
```

```python
#Cat.codes- dummies alternative

obj_df["studio"] = obj_df["studio"].astype('category')
obj_df.dtypes
clean["studio"] = obj_df["studio"].cat.codes

obj_df["genre"] = obj_df["genre"].astype('category')
obj_df.dtypes
clean["genre"] = obj_df["genre"].cat.codes

obj_df["broadcast"] = obj_df["broadcast"].astype('category')
obj_df.dtypes
clean["broadcast"] = obj_df["broadcast"].cat.codes

#obj_df["status"] = obj_df["status"].astype('category')
#obj_df.dtypes
#clean["status"] = obj_df["status"].cat.codes

obj_df["rating"] = obj_df["rating"].astype('category')
obj_df.dtypes
clean["rating"] = obj_df["rating"].cat.codes

#obj_df['duration'] = obj_df['duration'].astype('category')
#obj_df.dtypes
#clean['duration'] = obj_df['duration'].cat.codes

#obj_df["aired"] = obj_df["aired"].astype('category')
#obj_df.dtypes
#clean["aired"] = obj_df["aired"].cat.codes

obj_df["source"] = obj_df["source"].astype('category')
obj_df.dtypes
clean["source"] = obj_df["source"].cat.codes

obj_df["type"] = obj_df["type"].astype('category')
obj_df.dtypes
clean["type"] = obj_df["type"].cat.codes

obj_df.info()
#dummies
clean2=pd.get_dummies(clean.status)
```

```python
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

classifiers = [
    KNeighborsClassifier(3),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier()
    ]
for classifier in classifiers:
    model = classifier.fit(X_train, y_train)
    print("\n The Classifier is \n " +str(classifier))
    print("\n The Model's Score is %.3f" % model.score(X_test, y_test))
    y_pred_best = model.predict(X_test)
    cm=confusion_matrix(y_test, y_pred_best, labels=None, sample_weight=None)
    print(cm)
    print(classification_report(y_test,y_pred_best))
    sns.heatmap(pd.DataFrame(classification_report).iloc[:-1, :].T, annot=True)
```
