```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
AppleStore = pd.read_csv(r'C:\Users\\97252\OneDrive\שולחן העבודה\AppleStore.csv')
```

```python
#Using Seaborn
import seaborn as sns
```

```python
#Number of apps by genres
#1 sns

sns.set(font_scale=1.3)  # crazy big

plt.figure(figsize=(10, 10))

axes = sns.countplot(y=AppleStore.prime_genre)

csfont = {'fontname':'Comic Sans MS'}

axes.set_title("Apps quantity by genres",**csfont,fontsize=25)
axes.set_ylabel('genre',fontsize=20,**csfont)
axes.set_xlabel(' Number of Apps ',fontsize=20,**csfont)
plt.show()

plt.savefig(save_results_to + 'Number of apps by genres.png', dpi = 300)
```

```python
#Number of apps by categories
#2 sns

sns.set(font_scale=1.3)  # crazy big

plt.figure(figsize=(10, 10))

axes = sns.countplot(y=AppleStore.price)

csfont = {'fontname':'Comic Sans MS'}

axes.set_title("price Distribution of Apps",**csfont,fontsize=25)
axes.set_ylabel('Price by $',fontsize=20,**csfont)
axes.set_xlabel(' Number of Apps ',fontsize=20,**csfont)
plt.show()

plt.savefig(save_results_to + 'Number of apps by categories.png', dpi = 300)
```

```python
#3 sns

#lang.num distrbution by prime genres

g = sns.catplot(x="lang.num", y="prime_genre",
                data=aS, kind="strip",
                height=20, aspect=.7);
g.fig.suptitle('Language num by genres', y=1)

plt.savefig(save_results_to + 'lang.num distrbution by prime genres.png', dpi = 300)
```

```python
#4 sns

#Apps' Price Distribution by Genres

sns.set(font_scale=1.3)  # crazy big

plt.figure(figsize=(15,10))

axes = sns.stripplot(x="price", y="prime_genre", data=aS)
csfont = {'fontname':'Comic Sans MS'}

axes.set_title("Apps' Price Distribution by Genres",**csfont,fontsize=25)
axes.set_ylabel('Genre',fontsize=20,**csfont)
axes.set_xlabel(' Price $ ',fontsize=20,**csfont)


###############################how to save the graphs as a file#####################
plt.savefig(save_results_to + 'Apps Price Distribution by Genres.png', dpi = 300)
```

```python
#5 sns

#Pairplot of games

games=AppleStore[AppleStore['prime_genre']=='Games']

#pairplot of free games

freeGames=games[games['price']==0]

plotFree = sns.pairplot(freeGames)
plt.show()

plt.savefig(save_results_to + 'Pairplot of free games.png', dpi = 300)
```

```python
#6 sns

#Pairplot of paid games

paidGames=games[games['price']!='0']

plotPaid = sns.pairplot(paidGames)
plotPaid

plt.savefig(save_results_to + 'Pairplot of paid games.png', dpi = 300)
```

```python
#bonus- sns7

g = sns.catplot(x="user_rating", y="price",
                data=aS, kind="strip",
                height=20, aspect=.7);
g.fig.suptitle('user rating of Apps by price', y=1)

#shows that there isn't a real connection between price&score

plt.savefig(save_results_to + 'sns user rating of Apps by price.png', dpi = 300)
```
